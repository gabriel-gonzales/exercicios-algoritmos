## Cifra de Cesár

A Cifra de César é uma técnica de criptografia antiga, nomeada após Júlio César, que a utilizava para proteger comunicações militares. É uma cifra de substituição onde cada letra do texto original é substituída por outra que se encontra um número fixo de posições à frente no alfabeto. Por exemplo, com um deslocamento de 3, 'A' vira 'D', 'B' vira 'E', e assim por diante, reiniciando o ciclo após 'Z'. Hoje essa cifra é considerada simples e vulnerável a ataques de força bruta, mas, ainda assim, é um bom exemplo introdutório de cifras baseadas em substituição.

A entrada deste programa consiste em uma string e um número inteiro, cada um em uma linha. Faça um programa que faça o seguinte:

    Leia a primeira linha, que contém uma string a ser codificada. Esta string pode conter caracteres a-z, além de caracteres de pontuação e espaços;
    Leia a segunda linha, que contém um número inteiro representando o deslocamento para a Cifra de César;
    Aplique a Cifra de César na string lida, deslocando cada letra pelo número fornecido. As letras devem ser deslocadas ciclicamente, ou seja, após 'z' vem 'a' e após 'Z' vem 'A'. Os caracteres de pontuação e espaços devem permanecer inalterados;
    Exiba a string codificada.


```python
def cifra_de_cesar(texto, deslocamento):
    resultado = []

    for char in texto:
        if char.isalpha():
            base = ord('a') if char.islower() else ord('A')
            novo_char = chr((ord(char) - base + deslocamento) % 26 + base)
            resultado.append(novo_char)
        else:
            resultado.append(char)
    
    return ''.join(resultado)


texto = input().strip()
deslocamento = int(input().strip())

texto_codificado = cifra_de_cesar(texto, deslocamento)

print(texto_codificado)
```

## Busca em Vetor

Considere a seguinte entrada de um programa:

    1. Uma linha com um número inteiro N;
    2. N linhas, cada uma com um número inteiro;
    3. Uma linha com um número inteiro M;
    4. M linhas, cada uma com um número inteiro.

 

Escreva um programa que processe a entrada acima. Você deve armazenar as N linhas (item 2 da lista) em um vetor. Para cada número X das últimas M linhas (item 4 da lista), você deve escrever "Presente" caso o número X esteja no vetor, e "Ausente" caso contrário.

```python
N = int(input())

vetor = []

for i in range(N):
  numero = int(input())
  vetor.append(numero)

M = int(input())

for i in range(M):
  numero = int(input())
  if numero in vetor:
    print("Presente")
  else:
    print("Ausente")
  
```

Sua tarefa é escrever o arquivo funcoes.py para que o código acima funcione corretamente.

```python
def criaMatrizVazia(tamanho):
    matriz = []
    for i in range(tamanho):
        linha = []
        for j in range(tamanho):
            linha.append(0)
        matriz.append(linha)
    return matriz

def criaEmbarcacao(matriz, l, c, t, d):

  matriz[l][c] = 1 #posições matriz com 1

  if d == "V": # direção vertical
    for i in range(1, t): #tamanho da embarcação
      matriz[l + i][c] = 1 #posição da embarcação

  elif d == "H": # direção horizontal
    for i in range(1, t): 
      matriz[l][c + i] = 1
  return matriz

def mostraMatriz(matriz, tamanho):

  for i in range(tamanho):
    for j in range(tamanho):
      if matriz[i][j] == 0:
        print(".", end="") # fazendo com que 0 seja representado por '.'
      elif matriz[i][j] == 1:
        print("x", end="") # fazendo com que 1 seja representado por 'x'
      else:
        print("o", end="") # fazendo com que 2 seja representado por 'o'
    print() # para o print já criar a linha dps de cada linha da matriz


def acertou(matrizOponente, matrizJogador, l, c):

  if matrizJogador[l][c] == 0:
    if matrizOponente[l][c] == 1:  # quando acerta
      matrizJogador[l][c] = 1
      return True
    else:  # quaando o tiro vai na água
      matrizJogador[l][c] = 2
      return False
  else:  # caso já tenha sido atacado
    return False

def ganhou(matrizOpontente, matrizJogador, tamanho):
  for i in range(tamanho):
    for j in range(tamanho):
      if matrizOpontente[i][j] == 1 and matrizJogador[i][j] != 1:
        return False # pq alguns navios ainda não afundaram
  return True # no caso de todos já terem afundados
```

## Soma em Vetor com Pesos

Considere a seguinte entrada de um programa:

    Uma linha com um número inteiro N;
    N linhas, cada uma com um número inteiro;
    Uma linha com um número inteiro M;
    M linhas, cada uma com um número inteiro.

 

Escreva um programa que processe a entrada acima. Você deve armazenar as N linhas (item 2 da lista) em um vetor (e este deve ser o único vetor do código). Os números nas últimas M linhas (item 4 da lista) representam índices do vetor. Para cada uma dessas M linhas, caso o valor do índice seja inválido,você deve escrever "Indice X invalido", em que X é o índice lido da entrada. Um índice é válido se está no intervalo entre 0 e N-1. Você deve calcular a soma a seguir: para cada índice, se o valor do vetor correspondente a ele for par, ele deve ter peso 2, caso contrário, deve ter peso 3. Calcule essa soma ponderada e a escreva no final do programa.

```python
N = int(input())

vetor = []

for i in range(N):
  numero = int(input())
  vetor.append(numero)

M = int(input())

indices = []
soma = []
resultados = []

for i in range(M):
  indice = int(input())
  indices.append(indice)
  acesso = vetor[indice]
  resultados.append(str(acesso))
  soma += [acesso]

print(" + ".join(resultados) + " = " + str(sum((soma))))
```

## Atualização de Vetor

O programa abaixo inicia um vetor vazio e depois faz as seguintes operações:

    insere X: insere o inteiro X no final do vetor;
    remove Y: remove o elemento do vetor na posição Y;
    troca X Y: troca os valores dos elementos do vetor nas posições X e Y;
    presente X: verifica se o valor X está presente no vetor;
    sair: encerra o programa.

A fim de executar essas operações, o programa usa funções armazenadas no arquivo funcoes.py. A cada operação que alterar o vetor, o programa imprime seu conteúdo na saída.

    from funcoes import *

    linha = input()
    vetor = []

    while linha != "sair":
        operacao = linha.split()[0]
        match operacao:
            case "insere":
                valor = int(linha.split()[1])
                insere(vetor, valor)
                print(vetor)
            case "remove":
                posicao = int(linha.split()[1])
                remove(vetor, posicao)
                print(vetor)
            case "troca":
                posicao1 = int(linha.split()[1])
                posicao2 = int(linha.split()[2])
                troca(vetor, posicao1, posicao2)
                print(vetor)
            case "presente":
                valor = int(linha.split()[1])
                if presente(vetor, valor):
                        print("Valor " + str(valor) + " esta presente no vetor")
                else:
                        print("Valor " + str(valor) + " esta ausente no vetor")
        linha = input()

 
 

Sua tarefa é escrever o arquivo funcoes.py para que o código acima funcione. Esse é o arquivo que deve ser entregue.

```python
def insere(vetor, valor):
  vetor.append(valor) # insere o valor no vetor

def remove(vetor, posicao):

  if 0 <= posicao < len(vetor): # faz a verificação de que o indice seja válido pra lista
    vetor.pop(posicao) # remove o valor do vetor

def troca(vetor, posicao1, posicao2):

  if 0 <= posicao1 < len(vetor) and 0 <= posicao2 < len(vetor): # fazendo a mesma verificação do remove

    vetor[posicao1], vetor[posicao2] = vetor[posicao2], vetor[posicao1]

def presente(vetor, valor):

  return valor in vetor # verifica e retorna o True caso condição verdadeira
```
